```python
dataset = pd.read_csv('amazon_jobs_dataset.csv')
dataset.head()
```

```python
if sum(dataset.isnull().sum()) > 0: # There exist some NaN values
    dataset.dropna(inplace=True)
```

```python
stopwords = corpus.stopwords.words()
stopwords.extend(["years", "experience", "related", "field"]) # More words that clutter our results
```

```python
import re

qualifications = dataset['BASIC QUALIFICATIONS'].tolist()+dataset['PREFERRED QUALIFICATIONS'].tolist()
# joining the list to a single string and lower case the letter
qualifications_string = "".join(re.sub('[·,-/’()]', '', str(v)) for v in qualifications).lower()
```

```python
qualifications[10]

languages_list = ['Java', 'Python','C#', 'Mysql', 'hadoop', 'big data', 'pandas' 'data science', 'kafka', 'Ruby', 'Objective' , 'PHP','C', 'C++', 'R', 'MATLAB', 'bash','perl' ]
languages_list
```

```python
wordcount = dict((x.lower(),0) for x in languages_list)
print(wordcount)
```

```python
for w in re.findall(r"[[\w'+#-]+|[.!?;’]", qualifications_string):

    if w.lower() in wordcount.keys():
        wordcount[w] += 1
# print
print(wordcount)
```

```python
programming_language_popularity = sorted(wordcount.items(), key=lambda v: v[1], reverse=True)
programming_language_popularity
```

```python
df_popular_programming_lang = pd.DataFrame(programming_language_popularity,columns=['Language','Popularity'])
```

```python
df_popular_programming_lang = df_popular_programming_lang.set_index('Language')
df_popular_programming_lang.plot.bar()
```

```python
# plot
df_popular_programming_lang.plot.bar(x='Language',y='Popularity',figsize=(5,5),legend=False)
```

```python
basic_qualifications_string = "".join(re.sub('[·,-/’()]', '', str(v)) for v in qualifications)
```

```python
qualifications[0]
```

```python
degree_list = ["BA", "BS", "Bachelor's", "PhD","MS","Master's", 'Diploma', 'Associates']
```

```python
wordcount = dict((x,0) for x in degree_list)
for w in re.findall(r"[\w']+|[.,!?;’]", basic_qualifications_string):
    if w in wordcount:
        wordcount[w] += 1
# print
print(wordcount)
```

```python
degree_wanted = sorted(wordcount.items(), key=lambda kv: kv[1], reverse=True)
```

```python
df_degree_popular = pd.DataFrame(degree_wanted,columns=['Degree','Popularity'])

df_degree_popular.set_index('Degree')
```

```python
#df_degree_popular.plot.pie(y= subplots=True)
df_degree_popular.plot.bar(x='Degree',y='Popularity',figsize=(5,3), stacked=False)
```

```python
from collections import defaultdict

years = defaultdict(lambda: 0)

for w in re.findall(r'([0-9]) ', basic_qualifications_string):
     years[w] += 1
        
print(years)
```

```python
type(years)
years.items()
```

```python
years = sorted(years.items(), key=lambda kv: kv[1], reverse=True)
years
```

```python
df_years = pd.DataFrame(years,columns=['Years of experience','Popularity'])
df_years = df_years[::-1]
df_years
```

```python
df_years.plot.bar(x='Years of experience',y='Popularity',figsize=(10, 8), legend=False,stacked=True)
# add a suptitle
plt.title("Years of experiences needed for Amazon Jobs", fontsize=18)
# set xlabel to ""
plt.xlabel("Popularity", fontsize=14)
plt.ylabel("Years of experiences",fontsize=18)
# change xticks fontsize to 14
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
# finally show the plot
plt.show()
```

```python
threshold = 10
dataset = pd.read_csv('amazon_jobs_dataset.csv')
dataset.head()

location_value_counts = dataset.location.value_counts()
to_remove = location_value_counts[location_value_counts <= threshold].index
dataset['location'].replace(to_remove, np.nan, inplace=True)
# to_remove
```

```python

location_value_counts = dataset.location.value_counts()
location_value_counts = location_value_counts[::-1]

location_value_counts.plot.barh(figsize=(15, 15))
# add a suptitle
plt.title("Amazon Jobs Location", fontsize=24)
# set xlabel to ""
plt.xlabel("Popularity", fontsize=20)
plt.ylabel("Location",fontsize=20)
# change xticks fontsize to 14
plt.xticks(fontsize=24)
plt.yticks(fontsize=20)
# finally show the plot
plt.show()
```
